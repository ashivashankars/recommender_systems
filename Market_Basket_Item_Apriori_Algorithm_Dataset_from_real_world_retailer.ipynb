```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('train_data.csv', header=None)
print(store_data.head())
```

```python
# Group data by 'order_id' and collect 'product_name' into lists
grouped_data = store_data.groupby(0)[2].apply(list).reset_index(name='products')

# Convert the grouped data into a list of lists for Apriori
records = grouped_data['products'].tolist()

# Print the first few records to verify the format
print(records[:5])
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.1,min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(association_results)
#print(len(association_rules))
#print(association_rules[0])
print("items\n")
```

```python
if association_results:
  for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]

    # Check if there are at least two items before trying to print the rule
    if len(items) >= 2:
      print("Rule: " + items[0] + " -> " + items[1])
    else:
      # Handle cases with single items or other rule structures if necessary
      print("Rule: " + str(items) + " (Single item or unexpected format)")


    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list
    # Check if there are ordered statistics before trying to access them
    if len(item[2]) > 0 and len(item[2][0]) > 0:
        print("Confidence: " + str(item[2][0][2]))
        print("Lift: " + str(item[2][0][3]))
    else:
        print("Confidence: N/A")
        print("Lift: N/A")

    print("------------------------------------")
else:
  print("No association rules found with the given parameters.")
```

```python
test_data = pd.read_csv('test_data.csv', header=None)
print(test_data.head())
```

```python
# Group data by 'order_id' and collect 'product_name' into lists
# Skip the header row by starting from the second row (index 1)
grouped_test_data = test_data.iloc[1:].groupby(0)[1].apply(list).reset_index(name='products')

# Convert the grouped data into a list of lists for prediction
test_records = grouped_test_data['products'].tolist()

# Print the first few records to verify the format
print(test_records[0:5])
```

```python
def predict_next_item(basket, rules):
    """
    Predicts the next item in a basket based on association rules.

    Args:
        basket (list): A list of items in the current basket.
        rules (list): A list of association rules (from the apriori result).

    Returns:
        str or None: The predicted next item, or None if no rule matches.
    """
    predicted_item = None
    highest_confidence = 0

    for rule in rules:
        # Ensure rule has the expected structure before accessing elements
        if len(rule) >= 3 and len(rule[2]) > 0 and len(rule[2][0]) >= 2:
            antecedent = set(rule[2][0][0])
            consequent = set(rule[2][0][1])
            confidence = rule[2][0][2]

            # Check if the antecedent is a subset of the basket
            if antecedent.issubset(set(basket)):
                # If multiple rules match, choose the one with the highest confidence
                if confidence > highest_confidence:
                    # Return the first item in the consequent (assuming single item consequent)
                    if consequent: # Check if consequent is not empty
                        predicted_item = list(consequent)[0]
                        highest_confidence = confidence

    return predicted_item
```

```python
predictions = [predict_next_item(basket, association_results) for basket in test_records]
```
